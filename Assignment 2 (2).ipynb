# Question 1

```python
import numpy as np 
import pandas as pd 
import requests
```

```python
#reading URl into list
dfweb2=pd.read_html("http://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M", match='.+', flavor=None, header=0, index_col=None)
```

```python
#list loaded
dfweb2[0]
```

```python
#loading list into dataframe
dfnew4 =pd.DataFrame(dfweb2[0])
```

```python
dfnew4.info()
```

```python
dfnew4.head(10)
```

```python
dfnew4.rename(columns = {'Postcode':'PostalCode'}, inplace = True)
```

```python
#removing Borough which are not assigned
dfnew4.drop(dfnew4[dfnew4['Borough']=="Not assigned"].index,axis=0, inplace=True)
dfnew4.head()
```

```python
dfnew5 = dfnew4.reset_index()
dfnew5.head()
```

```python
dfnew5.shape
```

```python
#combining neighbourhood with postal code
dfnew6=dfnew5.groupby('PostalCode').agg(lambda x: ','.join(x))
dfnew6.head()
```

```python
#assigning borough to neighbourhood which are not assigned
dfnew6.loc[dfnew6['Neighbourhood']=="Not assigned",'Neighbourhood']=dfnew6.loc[dfnew6['Neighbourhood']=="Not assigned",'Borough']
```

```python
dfnew6.head()
```

```python
dfnew8 = dfnew6.reset_index()
dfnew8.head()
```

```python
dfnew8['Borough']= dfnew8['Borough'].str.replace('nan|[{}\s]','').str.split(',').apply(set).str.join(',').str.strip(',').str.replace(",{2,}",",")
```

```python
dfnew8.head()
```

```python
dfnew8.shape
```

```python
dfnew8.info()
```
